Matplotib basics
===

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import patheffects
import datetime
```

Minimal example plot ("object-oriented (OO) style") https://matplotlib.org/stable/users/getting_started/

```python
# "data"
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()

ax.plot(x, y)

print(type(fig))

print(type(ax))

plt.show()
```

Minimal example with pyplot-style

```python
plt.plot(x, y)

print(type(plt.gcf()))

print(type(plt.gca()))

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_subplot()

ax.plot(x, y)

plt.show()
```

```python
fig, axs = plt.subplots(
    2, 2,
    figsize=(5, 5),
    constrained_layout=True,
)

for ax in axs.flat:
    ax.plot(np.random.randn(100))
```

```python
fig, axs = plt.subplot_mosaic(
    [
        ['A', 'C'],
        ['B', 'C'],
        ['D', 'D']
    ],
    constrained_layout=True
)

for ax in axs.values():
    ax.plot(np.random.randn(100))
```
